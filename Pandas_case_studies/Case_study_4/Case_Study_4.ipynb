# Customer Nodes Exploration
- 1.How many unique nodes are there on the Data Bank system?
- 2.What is the number of nodes per region?
- 3.How many customers are allocated to each region?
- 4.How many days on average are customers reallocated to a different node?
- 5.What is the median, 80th and 95th percentile for this same reallocation days metric for each region?


```python
# 1.How many unique nodes are there on the Data Bank system?
customer_nodes['node_id'].nunique()
```

```python
# 2.What is the number of nodes per region?
# Merge the DataFrames on the 'region_id' column
merged_df = pd.merge(customer_nodes, regions, on='region_id', how='inner')

# Group by 'region_id' and 'region_name', then calculate the count of 'node_id' in each group
result_df = merged_df.groupby(['region_id', 'region_name']).size().reset_index(name='node_count')

# Display the result DataFrame
result_df
```

```python
# 3.How many customers are allocated to each region?

# Merge the DataFrames on the 'region_id' column
merged_df = pd.merge(customer_nodes, regions, on='region_id', how='inner')

# Group by 'region_id' and 'region_name', then calculate the count of distinct 'customer_id' in each group
result_df = merged_df.groupby(['region_id', 'region_name'])['customer_id'].nunique().reset_index(name='customer_count')

# Display the result DataFrame
result_df
```

```python
# 4.How many days on average are customers reallocated to a different node?

# Filter out rows where 'end_date' is not '9999-12-31'
filtered_df = customer_nodes[customer_nodes['end_date'] != '9999-12-31']

# Calculate the difference between 'end_date' and 'start_date' and calculate the average
avg_days = (filtered_df['end_date'] - filtered_df['start_date']).mean().days

# Display the result
print(f"Average number of days: {avg_days}")
```

```python
# 5.What is the median, 80th and 95th percentile for this same reallocation days metric for each region?

# Calculate median, 80th, and 95th percentiles for each region
result_summary = merged_df.groupby('region_id')['reallocation_days'].agg([
    ('median', 'median'),
    ('80th_percentile', lambda x: x.quantile(0.8)),
    ('95th_percentile', lambda x: x.quantile(0.95))
]).reset_index()

# Merge with regions DataFrame to include region_name
result_summary = pd.merge(result_summary, regions[['region_id', 'region_name']], on='region_id')

# Display the result
result_summary[['region_id', 'region_name', 'median', '80th_percentile', '95th_percentile']]
```

# Customer Transactions
- 1.What is the unique count and total amount for each transaction type?
- 2.What is the average total historical deposit counts and amounts for all customers?
- 3.For each month-how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?
- 4.what is the closing balance for each customer at the end of the month?


```python
# 1.What is the unique count and total amount for each transaction type?

# Group by 'txn_type' and calculate unique count and total amount
result_df = customer_transaction.groupby('txn_type').agg(
    unique_count=('customer_id', 'nunique'),
    total_amount=('txn_amount', 'sum')
).reset_index()

# Display the result
result_df
```

```python
# 2.What is the average total historical deposit counts and amounts for all customers? 

# Filter rows with txn_type 'deposit'
deposit_data = customer_transaction[customer_transaction['txn_type'] == 'deposit']

# Calculate average total historical deposit counts
average_total_deposit_count = deposit_data.groupby('customer_id').size().mean()

# Calculate average total historical deposit amounts
average_total_deposit_amount = deposit_data.groupby('customer_id')['txn_amount'].sum().mean()

# Display the results
print("Average Total Historical Deposit Counts:", round(average_total_deposit_count))
print("Average Total Historical Deposit Amounts:", f"${round(average_total_deposit_amount, 2)}")
```

```python
#3.For each month-how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?


# Extract month, and count transactions for deposit, withdrawal, and purchase
transaction_count_per_month = customer_transaction.groupby(['customer_id', customer_transaction['txn_date'].dt.month]).agg(
    deposit_count=('txn_type', lambda x: (x == 'deposit').sum()),
    withdrawal_count=('txn_type', lambda x: (x == 'withdrawal').sum()),
    purchase_count=('txn_type', lambda x: (x == 'purchase').sum())
).reset_index()

# Filter customers who made more than 1 deposit and either 1 purchase or 1 withdrawal in a single month
filtered_data = transaction_count_per_month[(transaction_count_per_month['deposit_count'] > 1) &
                                            ((transaction_count_per_month['purchase_count'] == 1) |
                                             (transaction_count_per_month['withdrawal_count'] == 1))]

# Count customers for each month
result_df = filtered_data.groupby('txn_date').size().reset_index(name='customer_count')

# Display the result DataFrame
result_df
```

```python
# 4.what is the closing balance for each customer at the end of the month?
# Extract month, calculate net_transaction_amt, and sort by customer_id
txn_monthly_balance_df = customer_transaction.assign(
    txn_month=customer_transaction['txn_date'].dt.month,
    net_transaction_amt=customer_transaction.apply(
        lambda row: row['txn_amount'] if row['txn_type'] == 'deposit' else -row['txn_amount'], axis=1
    )
).groupby(['customer_id', 'txn_month']).agg(
    net_transaction_amt=('net_transaction_amt', 'sum')
).reset_index().sort_values(by=['customer_id', 'txn_month'])

# Calculate closing_balance using cumsum within each customer_id group
txn_monthly_balance_df['closing_balance'] = txn_monthly_balance_df.groupby('customer_id')['net_transaction_amt'].cumsum()

# Display the result DataFrame
txn_monthly_balance_df[['customer_id', 'txn_month', 'net_transaction_amt', 'closing_balance']]
```
